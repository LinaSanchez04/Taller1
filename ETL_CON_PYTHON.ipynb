```python
# Transform: realizar algunas operaciones de procesamiento de datos
df['promedio'] = (df[' "1958"'] + df[' "1959"'] + df[' "1960"'])/3

print(df)

df = df[df[' "1960"'] > 450]
df = df.dropna()
df = df.rename(columns={' "1958"': 'año 1958', ' "1959"': 'año 1959', ' "1960"': 'año 1960'})

print(df)

df = df.rename(columns={'Month':'Mes'})

print(df)
```
